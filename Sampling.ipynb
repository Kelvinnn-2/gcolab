```python
filename = "/content/drive/MyDrive/ColabNotebooks/dataset/1000000 Sales Records.csv" 
n = sum(1 for line in open(filename))-1  # Calculate number of rows in file
s = n//10  # sample size of 10%
skip = sorted(rnd.sample(range(1, n+1), n-s))  # n+1 to compensate for header 
df = pd.read_csv(filename, skiprows=skip)
df
```

```python
datafile = "/content/drive/MyDrive/ColabNotebooks/dataset/1000000 Sales Records.csv"
result = None
for chunk in pd.read_csv("/content/drive/MyDrive/ColabNotebooks/dataset/1000000 Sales Records.csv", chunksize=1000):
    voters_street = chunk[
        "Region"]
    chunk_result = voters_street.value_counts()
    if result is None:
        result = chunk_result
    else:
        result = result.add(chunk_result, fill_value=0)

result.sort_values(ascending=False, inplace=True)
print(result)
```
